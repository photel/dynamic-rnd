```python
def mann_whitney_u_test(values_model_a, values_model_b, alpha=0.05):
    '''
        Perform the Mann-Whitney U test to compare two DRL models.

        Parameters:
            - values_model_a: Values from Model A across different seeds.
            - values_model_b: Values from Model B across different seeds.
            - alpha: Significance level

        Returns:
            - test_statistic: The Mann-Whitney U test statistic
            - p_value: The p-value of the test
            - significant: A boolean - True if the difference is statistically significant
    '''
    # Combine values from all seeds for each model
    combined_values_a = np.concatenate(values_model_a)
    combined_values_b = np.concatenate(values_model_b)

    # Perform Mann-Whitney U test
    U, p_value = mannwhitneyu(combined_values_a, combined_values_b, alternative='greater')

    # Determine if the result is significant
    significant = p_value < alpha

    return U, p_value, significant
```

```python
def display_results(test_statistic, p_value, significant, test_type):
    print(f'{test_type}: {test_statistic:.2f}')
    print(f'P-value: {p_value:.4f}')
    if significant:
        print(f'The difference is significant')
    else:
        print(f'The difference is NOT significant')
```

```python
def cohen_d(sample1, sample2):
    # Calculate the means and standard deviations
    mean1, mean2 = np.mean(sample1), np.mean(sample2)
    std1, std2 = np.std(sample1, ddof=1), np.std(sample2, ddof=1)

    # Calculate the average standard deviation
    s_avg = np.sqrt((std1**2 + std2**2) / 2)

    # Calculate Cohen's d
    d = (mean1 - mean2) / s_avg
    return d

d_rewards = cohen_d(d_rnd_final_means, rnd_final_means)
d_rooms = cohen_d(d_rnd_room_final_means, rnd_room_final_means)

print(f"Cohen's d rewards: {d_rewards}")
print(f"Cohen's d rooms: {d_rooms}")
```
